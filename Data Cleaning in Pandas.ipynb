# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\Data Analytics\Data\Customer Call List.xlsx")
df
```

```python
# specify n to indicate how many times it splits
# Creates three extra columns, appends it, doesn't replace Address column
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand=True)
df
```

```python
# .replace with ^ = replaces any character NOT a-z, A-Z, 0-9 with a blank space '', ^ = not
# regex = for pandas to interpret the replacement as a regular expression replacement
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
df
```

```python
# [0:3] = 0,1,2 [3:6] = 3,4,5 [6:10] = 6,7,8,9
# df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) does not work because you need to change to str first
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x:str(x)) assign to variable before inputting next code to ensure change sticks
# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) test if it works correctly before assigning it to variable
# df["Phone_Number"].str.replace('nan--','') 
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')
df
```

```python
# make sure to reassign to variable to save changes
df = df.drop_duplicates()
df
```

```python
# can also do inplace = True to also save it, but for convenience you reassign the 
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# if you don't want strip to apply to everything, then specify column name
# can specify left or right side by doing either rstrip or lstrip, strip applies both
# default strip eliminates white space
# More Complicated way
# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")

# 'Right Way'
df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df

# can also use
# df['Do_Not_Contact'] = df['Do_Not_Contact'].replace({'N': 'No', 'Y': 'Yes'})
```

```python
df = df.replace('N/a','')
# df = df.replace('NaN','') does not work because NaN is null, not a str
df = df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x,"Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
df

# or if you don't want to loop
# df = df.loc[df["Do_Not_Contact"] == "N"] to filter out only "N"
# or df = df[df["Do_Not_Contact"] != "Y"]
```

```python
for x in df.index:
    if df.loc[x,"Phone_Number"] == '':
        df.drop(x, inplace=True)
df

# Another way to drop null values --> df.dropna(subset="Phone_Number"), inplace=True)
```

```python
# drop=True for not keeping the original index
df = df.reset_index(drop=True)
df
```
