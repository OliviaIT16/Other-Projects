# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Data Analytics\Data\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
# default is a line plot
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
# can change look of plot
print(plt.style.available)
plt.style.use('bmh')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Rating', ylabel = 'Scores')
```

```python
# you can specify what column you want by doing ['Flavor Rating'], there's no legend because it appears when there are multiple values
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
df.plot.barh(stacked = True)
```

```python
# s = size, c = color
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100, c = 'Yellow')
```

```python
df.plot.hist()
```

```python
# bins = represents how many ticks there are 
# 10 is default
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5))
```

```python
df.plot.pie(y='Flavor Rating', figsize=(10,6))
```
